Variables
---------

Python can broadly be divided in verbs and nouns: things which *do* things, and things which *are* things. In Python, the verbs can be *commands*, *functions*, or *methods*. We won't worry too much about the distinction here - suffice it to say, they are the parts of code which manipulate data, calculate values, or show things on the screen.

The simplest proper noun object in Python is the *variable*. Variables are given names and store information. This can be, for example, numeric, text, or boolean (true/false). These are all statements setting up variables:

n = 1

t = "hi"

b = True

Now let's try this in code:

```python
n = 1

t = "hi"

b = True
```

The value of this is we now have values tied to these variables - so every time we want to use it, we can refer to the variable:

```python
n
```

```python
t
```

Flow
----

Python, in common with all programming languages, executes commands in a sequence - we might refer to this as the "ineluctable march of the machines", but it's more common referred to as the *flow* of the code (we'll use the word "code" a lot - it just means commands written in the programming language). In most cases, code just executes in the order it's written. This is true within each *cell* (each block of text in the notebook), and it's true when we execute the cells in order; that's why we can refer back to the variables we defined earlier:

```python
print(n)
```

If we make a change to one of these variables, say n:

```python
n = 3
```

and execute the above "print n" command, you'll see that it has changed n to 3. So if we go out of order, the obvious flow of the code is confused. For this reason, try to write your code so it executes in order, one cell at a time. At least for the moment, this will make it easier to follow the logic of what you're doing to data.

```python
import pandas as pd
import requests
from io import StringIO

# GitHub URL（不改变）
url = 'https://github.com/newzealandpaul/Maritime-Pirate-Attacks/raw/main/data/csv/pirate_attacks.csv'

# 使用 requests 获取 CSV 文件内容
response = requests.get(url)

# 读取 CSV 数据
data = pd.read_csv(StringIO(response.text))

# 确保 'date' 列是日期时间格式
data['date'] = pd.to_datetime(data['date'])

# 过滤日期范围
filtered_data = data[(data['date'] >= '1993-01-01') & (data['date'] <= '1993-03-31')]

# 按季度统计数量
quarterly_counts = filtered_data.resample('Q', on='date').size()

# 转换索引为季度格式
quarterly_counts.index = quarterly_counts.index.to_period('Q').strftime('%Y Q%q')

# 输出结果
print(quarterly_counts)
```

```python
import pandas as pd
import requests
from io import StringIO

# GitHub URL（不改变）
url = 'https://github.com/newzealandpaul/Maritime-Pirate-Attacks/raw/main/data/csv/pirate_attacks.csv'

# 使用 requests 获取 CSV 文件内容
response = requests.get(url)
data = StringIO(response.text)

# 读取 CSV 文件
df = pd.read_csv(data)

# 确保 'date' 列为 datetime 类型
df['date'] = pd.to_datetime(df['date'])

# 过滤出1993 Q1到2020 Q4的数据
start_date = '1993-01-01'
end_date = '2020-12-31'
filtered_data = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

# 按季度重采样并计算每个季度的数据量
quarterly_counts = filtered_data.resample('Q', on='date').size()

# 转换索引为季度格式
quarterly_counts.index = quarterly_counts.index.to_period('Q').strftime('%Y Q%q')

# 输出结果
print(quarterly_counts)

```

```python
# Parallel Analysis for Aden Gulf and Southeast Asia:

import pandas as pd
import requests
from io import StringIO
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols

# Load data
url = 'https://github.com/newzealandpaul/Maritime-Pirate-Attacks/raw/main/data/csv/pirate_attacks.csv'
response = requests.get(url)
data = StringIO(response.text)
df = pd.read_csv(data)

# Ensure 'date' column is datetime
df['date'] = pd.to_datetime(df['date'])

# Filter data for the relevant time period
start_date = '1993-01-01'
end_date = '2008-12-31'  # Change to before the policy implementation
df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

# Define treatment and control groups
# Treatment group: Aden Gulf (Yemen and Somalia)
# Control group: Southeast Asia (specific countries)
df['treatment'] = np.where(df['nearest_country'].isin(['YEM', 'SOM', 'ERI', 'DJI']), 1,
                            np.where(df['nearest_country'].isin(['PHL', 'VNM', 'KHM', 'THA', 'MYS', 'IDN', 'SGP']), 0, np.nan))

# Remove rows where treatment is NaN (not in either group)
df = df.dropna(subset=['treatment'])

# Create quarter variable
df['quarter'] = df['date'].dt.to_period('Q')  # Create period-based quarter
agg_df = df.groupby(['quarter', 'treatment']).size().reset_index(name='num_attacks')

# Convert 'quarter' to string for plotting
agg_df['quarter'] = agg_df['quarter'].astype(str)

# Visualize the trends for treatment and control groups
plt.figure(figsize=(12, 6))
sns.lineplot(data=agg_df, x='quarter', y='num_attacks', hue='treatment', marker='o')
plt.title('Trends in Pirate Attacks Before Operation Atalanta (1993-2008)')
plt.ylabel('Number of Pirate Attacks')
plt.xlabel('Time (Quarter)')
plt.legend(title='Region (1=Aden Gulf, 0=Southeast Asia)')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()

# Run regression to test for parallel trends
# We can include a time trend variable to see if the slopes are significantly different
agg_df['time'] = np.arange(len(agg_df))  # Create a time variable for the regression

# Run the regression for the pre-policy period
pre_did_model = ols('num_attacks ~ treatment + time + treatment*time', data=agg_df).fit()
print(pre_did_model.summary())

# Conclusion based on the regression results
if pre_did_model.pvalues['treatment:time'] < 0.05:
    print("Therefore: The assumptions for a parallel analysis are not met.")
else:
    print("Therefore: The assumptions for a parallel analysis are met.")
```

```python
import pandas as pd
import requests
from io import StringIO

# GitHub URL（不改变）
url = 'https://github.com/newzealandpaul/Maritime-Pirate-Attacks/raw/main/data/csv/pirate_attacks.csv'

# 使用 requests 获取 CSV 文件内容
response = requests.get(url)
data = StringIO(response.text)

# 读取 CSV 文件
df = pd.read_csv(data)

# 确保 'date' 列为 datetime 类型
df['date'] = pd.to_datetime(df['date'])

# 定义几内亚湾国家列表（包括相应的 ISO 代码）
guinea_gulf_countries = ['NGA', 'BEN', 'TGO', 'GHA', 'CIV', 'CMR']

# 过滤出几内亚湾地区的数据
filtered_data_Guinea_Gulf = df[df['nearest_country'].isin(guinea_gulf_countries)]

# 确保仍然过滤到2000年到2020年范围内的数据
start_date = '2000-01-01'
end_date = '2020-12-31'
filtered_data_Guinea_Gulf = filtered_data_Guinea_Gulf[(filtered_data_Guinea_Gulf['date'] >= start_date) & (filtered_data_Guinea_Gulf['date'] <= end_date)]

# 计算每年海盗袭击数量
annual_counts = filtered_data_Guinea_Gulf.set_index('date').resample('Y').size()

# 过滤出2000到2020的数据
annual_counts_filtered = annual_counts[(annual_counts.index.year >= 2000) & (annual_counts.index.year <= 2020)]

# 输出从2000年到2020年每年的海盗袭击数量
print("Pirate Attacks in Gulf of Guinea from 2000 to 2020:")
total_attacks = 0
for year in range(2000, 2021):
    count = annual_counts_filtered.get(pd.Timestamp(f"{year}-12-31"), 0)  # 获取每年的数据
    total_attacks += count
    print(f"{year}: {count} attacks")

print(f"\nTotal attacks from 2000 to 2020: {total_attacks} attacks")
```

```python
# Parallel Analysis for Aden Gulf and Southeast Asia

import pandas as pd
import requests
from io import StringIO
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
import numpy as np # Import numpy

# GitHub URL for the cleaned piracy data
url = 'https://raw.githubusercontent.com/yuhengding1206/QM2-Project/main/cleaned_piracy_data_with_calculations.csv'

# Fetch the CSV file content
response = requests.get(url)
df = pd.read_csv(StringIO(response.content.decode('utf-8')))

# Select the desired columns
selected_data = df[['Year', 'Aden_per_attack', 'Southeast_asia_per_attack']]

# Filter for the specific years (1993 to 2020)
filtered_data = selected_data[(selected_data['Year'] >= 1993) & (selected_data['Year'] <= 2020)]

# Visualize the trends for Aden Gulf and Southeast Asia
plt.figure(figsize=(12, 6))
plt.plot(filtered_data['Year'], filtered_data['Aden_per_attack'], label='Aden per Attack', marker='o')
plt.plot(filtered_data['Year'], filtered_data['Southeast_asia_per_attack'], label='Southeast Asia per Attack', marker='o')

# Draw a vertical dashed line at the year 2009 (for reference)
plt.axvline(x=2009, color='red', linestyle='--', label='Atalanta Operation Start')

# Adding labels and title
plt.title('Pirate Attacks Over Time (1993-2020)')
plt.xlabel('Year')
plt.ylabel('Number of Attacks per Region')
plt.xticks(filtered_data['Year'])  # Set x-ticks to show each year
plt.legend()
plt.grid()

# Show the plot
plt.show()

# Prepare data for regression analysis to test for parallel trends
# Create a treatment variable (1 for Aden, 0 for Southeast Asia)
filtered_data['treatment'] = np.where(filtered_data['Aden_per_attack'].notnull(), 1, 0)

# Create a time variable for the regression
filtered_data['time'] = np.arange(len(filtered_data))

# Run the regression for the pre-policy period (1993-2008)
pre_did_model = ols('Aden_per_attack ~ treatment + time + treatment*time', data=filtered_data[filtered_data['Year'] < 2009]).fit()
print(pre_did_model.summary())

# Conclusion based on the regression results
if pre_did_model.pvalues['treatment:time'] < 0.05:
    print("Therefore: The assumptions for a parallel analysis are not met.")
else:
    print("Therefore: The assumptions for a parallel analysis are met.")
```

```python
# Differences in differences analysis for Aden Gulf and Southeast Asia:

# Differences in differences analysis for Aden Gulf and Southeast Asia:

import pandas as pd
import requests
from io import StringIO
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
import numpy as np
import seaborn as sns # Import seaborn

# GitHub URL for the cleaned piracy data
url = 'https://raw.githubusercontent.com/yuhengding1206/QM2-Project/main/cleaned_piracy_data_with_calculations.csv'

# Fetch the CSV file content
response = requests.get(url)
df = pd.read_csv(StringIO(response.content.decode('utf-8')))

# Select the desired columns for Aden Gulf and Guinea Gulf
selected_data = df[['Year', 'Aden_per_attack', 'Guinea_per_attack']]

# Filter for the specific years (2000 to 2020)
filtered_data = selected_data[(selected_data['Year'] >= 2000) & (selected_data['Year'] <= 2020)]

# Create treatment variable for Aden Gulf (1) and Guinea Gulf (0)
filtered_data['treatment'] = np.where(filtered_data['Aden_per_attack'].notnull(), 1, 0)

# Create post variable (1 for post-2008, 0 for pre-2009)
filtered_data['post'] = np.where(filtered_data['Year'] >= 2009, 1, 0)

# Create an interaction term for post and treatment
filtered_data['post_treatment'] = filtered_data['post'] * filtered_data['treatment']


# Create a variable for the counts of attacks: Aden Gulf is already counted in 'Aden_per_attack',
# for Guinea Gulf, we can reverse engineer the logic.
filtered_data['num_attacks'] = np.where(filtered_data['treatment'] == 1, filtered_data['Aden_per_attack'], filtered_data['Guinea_per_attack'])

# Run DID regression
# Changed 'agg_df' to 'filtered_data' to reference the correct DataFrame
did_model = ols('num_attacks ~ post + treatment + post_treatment', data=filtered_data).fit()
print(did_model.summary())

# Visualize boxplot for pre and post periods by treatment group
# Changed 'did' to 'filtered_data' and 'unemployment' to 'num_attacks'
filtered_data['category'] = filtered_data['treatment'].astype(str) + filtered_data['post'].astype(str)
sns.boxplot(x='category', y='num_attacks', hue='treatment', data=filtered_data).set_xticklabels(
    ["Pre x Treatment", "Pre x Control", 'Post x Treatment', 'Post x Control']
)
plt.xlabel('')
plt.title('Unemployment Rates by Treatment and Post Treatment')
plt.show()

```
