Variables
---------

Python can broadly be divided in verbs and nouns: things which *do* things, and things which *are* things. In Python, the verbs can be *commands*, *functions*, or *methods*. We won't worry too much about the distinction here - suffice it to say, they are the parts of code which manipulate data, calculate values, or show things on the screen.

The simplest proper noun object in Python is the *variable*. Variables are given names and store information. This can be, for example, numeric, text, or boolean (true/false). These are all statements setting up variables:

n = 1

t = "hi"

b = True

Now let's try this in code:

```python
n = 1

t = "hi"

b = True
```

Note that each command is on a new line; other than that, the *syntax* of Python should be fairly clear. We're setting these variables equal to the letters and numbers and phrases and booleans. **What's a boolean?**

The value of this is we now have values tied to these variables - so every time we want to use it, we can refer to the variable:

```python
n
```

```python
t
```

```python
b
```

Because we've defined these variables in the early part of the notebook, we can use them later on.

Where is my data?
-----------------

When we work in excel and text editors, we're used to seeing the data onscreen - and if we manipulate the data in some way (averaging or summing up), we see both the inputs and outputs on screen. They're there in the background, but it's usually worth checking in on them from time to time, to see whether our processes are doing what we think they're doing.

This is pretty easy to do - we can just type the variable name, or "print(*variable name*)":

```python
n = n+1
print(n)
print(t)
print(b)
```

Flow
----

Python, in common with all programming languages, executes commands in a sequence - we might refer to this as the "ineluctable march of the machines", but it's more common referred to as the *flow* of the code (we'll use the word "code" a lot - it just means commands written in the programming language). In most cases, code just executes in the order it's written. This is true within each *cell* (each block of text in the notebook), and it's true when we execute the cells in order; that's why we can refer back to the variables we defined earlier:

```python
print(n)
```

If we make a change to one of these variables, say n:

```python
n = 3
```

and execute the above "print n" command, you'll see that it has changed n to 3. So if we go out of order, the obvious flow of the code is confused. For this reason, try to write your code so it executes in order, one cell at a time. At least for the moment, this will make it easier to follow the logic of what you're doing to data.

```python
import pandas as pd
import requests
from io import StringIO

# GitHub URL（不改变）
url = 'https://github.com/newzealandpaul/Maritime-Pirate-Attacks/raw/main/data/csv/pirate_attacks.csv'

# 使用 requests 获取 CSV 文件内容
response = requests.get(url)

# 读取 CSV 数据
data = pd.read_csv(StringIO(response.text))

# 确保 'date' 列是日期时间格式
data['date'] = pd.to_datetime(data['date'])

# 过滤日期范围
filtered_data = data[(data['date'] >= '1993-01-01') & (data['date'] <= '1993-03-31')]

# 按季度统计数量
quarterly_counts = filtered_data.resample('Q', on='date').size()

# 转换索引为季度格式
quarterly_counts.index = quarterly_counts.index.to_period('Q').strftime('%Y Q%q')

# 输出结果
print(quarterly_counts)
```

```python
import pandas as pd
import requests
from io import StringIO
import matplotlib.pyplot as plt

# GitHub URL
url = 'https://github.com/newzealandpaul/Maritime-Pirate-Attacks/raw/main/data/csv/pirate_attacks.csv'

# 使用 requests 获取 CSV 文件内容
response = requests.get(url)
data = StringIO(response.text)

# 读取 CSV 文件
df = pd.read_csv(data)

# 确保 'date' 列为 datetime 类型
df['date'] = pd.to_datetime(df['date'])

# 过滤出1993 Q1到2020 Q4的数据
start_date = '1993-01-01'
end_date = '2020-12-31'

# 过滤出 YEM 和 SOM 的数据
filtered_data_yem_som = df[df['nearest_country'].isin(['YEM', 'SOM'])]
filtered_data_yem_som = filtered_data_yem_som[(filtered_data_yem_som['date'] >= start_date) & (filtered_data_yem_som['date'] <= end_date)]

# 按季度重采样并计算每个季度的数据量
quarterly_counts_yem_som = filtered_data_yem_som.resample('Q', on='date').size()

# 过滤出 PHL, VNM, KHM, THA, MYS, IDN 的数据
filtered_data_asia = df[df['nearest_country'].isin(['PHL', 'VNM', 'KHM', 'THA', 'MYS','IDN'])]
filtered_data_asia = filtered_data_asia[(filtered_data_asia['date'] >= start_date) & (filtered_data_asia['date'] <= end_date)]

# 按季度重采样并计算每个季度的数据量
quarterly_counts_asia = filtered_data_asia.resample('Q', on='date').size()

# 转换索引为季度格式
quarterly_counts_yem_som.index = quarterly_counts_yem_som.index.to_period('Q').strftime('%Y Q%q')
quarterly_counts_asia.index = quarterly_counts_asia.index.to_period('Q').strftime('%Y Q%q')

# 绘制折线图
plt.figure(figsize=(14, 7))
plt.plot(quarterly_counts_yem_som.index, quarterly_counts_yem_som.values, marker='o', linestyle='-', color='b', label='YEM & SOM')
plt.plot(quarterly_counts_asia.index, quarterly_counts_asia.values, marker='o', linestyle='-', color='r', label='PHL, VNM, KHM, THA, MYS, IDN')

plt.title('Number of Pirate Attacks from 1994 Q1 to 2020 Q4')
plt.xlabel('Time (Quarter)')
plt.ylabel('Number of Pirate Attacks')

# 设置x轴刻度，只保留每年的Q1
xticks = [f"{year} Q1" for year in range(1994, 2021)]
xtick_positions = [i for i, q in enumerate(quarterly_counts_yem_som.index) if 'Q1' in q]
plt.xticks(xtick_positions, xticks, rotation=45)  # 旋转x轴标签以便更好地显示

operation_atlanta_q1 = '2009 Q1'
plt.axvline(x=quarterly_counts_yem_som.index.get_loc(operation_atlanta_q1), color='black', linestyle='dashed', label='Operation Atalanta')

plt.grid()
plt.legend(title='Countries')
plt.tight_layout()  # 调整布局以避免标签重叠
plt.show()
```

```python
import pandas as pd
import requests
from io import StringIO
import numpy as np
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
url = 'https://github.com/newzealandpaul/Maritime-Pirate-Attacks/raw/main/data/csv/pirate_attacks.csv'
response = requests.get(url)
data = StringIO(response.text)
df = pd.read_csv(data)

# Ensure 'date' column is datetime
df['date'] = pd.to_datetime(df['date'])

# Filter data for the relevant time period
start_date = '1993-01-01'
end_date = '2020-12-31'
df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]

# Define treatment and control groups
df['treatment'] = np.where(df['nearest_country'].isin(['YEM', 'SOM']), 1, 0)

# Define 'post' variable: 1 if after Operation Atalanta (2009 Q1), 0 otherwise
df['post'] = np.where(df['date'] >= '2009-01-01', 1, 0)

# Create DID interaction term
df['post_treatment'] = df['post'] * df['treatment']

# Aggregate number of attacks by quarter and group
df['quarter'] = df['date'].dt.to_period('Q')  # Create period-based quarter
agg_df = df.groupby(['quarter', 'treatment', 'post', 'post_treatment']).size().reset_index(name='num_attacks')

# Convert 'quarter' to string for plotting
agg_df['quarter'] = agg_df['quarter'].astype(str)

# Visualize number of attacks over time by region
sns.lineplot(data=agg_df, x='quarter', y='num_attacks', hue='treatment', marker='o')
plt.axvline(x='2009Q1', color='black', linestyle='dashed', label='Operation Atalanta')
plt.title('Number of Pirate Attacks Before and After Operation Atalanta')
plt.ylabel('Number of Pirate Attacks')
plt.xlabel('Time (Quarter)')
plt.legend(title='Region (1=YEM&SOM, 0=Asia)')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()

# Run DID regression
did_model = ols('num_attacks ~ post + treatment + post_treatment', data=agg_df).fit()
print(did_model.summary())

# Visualize boxplot for pre and post periods by treatment group
agg_df['category'] = agg_df['treatment'].astype(str) + agg_df['post'].astype(str)
sns.boxplot(x='category', y='num_attacks', hue='treatment', data=agg_df).set_xticklabels(
    ['Pre x Control', 'Post x Control', 'Pre x Treatment', 'Post x Treatment']
)
plt.title('Number of Pirate Attacks by Treatment and Post Periods')
plt.xlabel('')
plt.ylabel('Number of Pirate Attacks')
plt.legend(title='Region')
plt.tight_layout()
plt.show()
```
